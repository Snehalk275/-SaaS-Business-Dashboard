```python
import pandas as pd

customers = pd.read_csv("customer.csv")
subscriptions = pd.read_csv("subscriptions.csv")
usage = pd.read_csv("usage.csv")
financials = pd.read_csv("financials.csv")
```

```python
df = customers.merge(subscriptions, on="customer_id") \
              .merge(usage, on="customer_id") \
              .merge(financials, on="customer_id")
```

```python
df.groupby('country')['profit'].sum().sort_values(ascending=False)   #Profit by country
```

```python
avg_logins = df['logins'].mean()

high_risk = df[
    (df['status'] == 'Active') &
    (df['logins'] < avg_logins) &
    (df['support_tickets'] > 5)
]
```
